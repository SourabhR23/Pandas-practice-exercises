## Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

## Load the data

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chip = pd.read_csv(url, sep = '\t')
```

## Top 10 enteries

```python
chip.head(10)
```

```python
# Clean the item_price and transform it in a float
prices = [float(value[1 : -1]) for value in chip['item_price']]
prices[0]
```

```python
# Reassign the column with cleaned prices
chip['item_price'] = prices
chip.head()
```

```python
# Delete the duplicates
chip_filtered = chip.drop_duplicates()
chip_filtered.head()
```

### Select only the products with quantity equals to 1

```python
quan_one = chip_filtered[chip_filtered['quantity'] == 1]
quan_one
```

### What is the price of each item?
Print a data frame with only two columns item_name and item_price

```python
# Deleting the duplicates
chip_filtered = chip.drop_duplicates(['quantity', 'item_name'])

# Products with quantity equals to 1
quan_one = chip_filtered[chip_filtered['quantity'] == 1]

# Creating new dataframe
price_each_item = quan_one[['item_name', 'item_price']]

# Sorting the quantity from most expensive to less expensive
price_each_item.sort_values('item_price', ascending = False)
```

### Sort by the name of the item

```python
sorted_item_name = chip.sort_values('item_name')
sorted_item_name.head(20)
```

### What was the quantity of the most expensive item ordered?

```python
chip.sort_values('item_price', ascending = False).head(1)
```

### How many times was a Veggie Salad Bowl ordered?

```python
salad_bowl = chip[chip['item_name'] == 'Veggie Salad Bowl']
len(salad_bowl)
```

### How many times did someone order more than one Canned Soda?

```python
soda_times = chip[(chip['item_name'] == 'Canned Soda') & (chip['quantity'] > 1)]
len(soda_times)
```
